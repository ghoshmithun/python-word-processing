```python
import pandas as pd
    
def create_one_year_or_cumm_data(excel_sheet_name):
    df=pd.read_excel("C:/Users/mithu/Desktop/mona study/RW 45 -ANNUAL CLOSING DATA  21-22 BANGALORE final.xlsx",skiprows=2,sheet_name=excel_sheet_name)
    df['category_new']=df['Category'].apply(lambda x: 'EWS' if x in ['EWS NEW','EWS OLD'] else 'MIG')
    df_pivot=df.groupby(['NW','BR CODE','category_new','Construction completed/ Under Construction']).agg({'SUBSIDY':['count','sum']}).reset_index()
    df_pivot.columns=['NW','BR CODE','category_new','Construction completed/ Under Construction','SUBSIDY_count','SUBSIDY_sum']
    return df_pivot

one_year_df=create_one_year_or_cumm_data('RW-45-I-E(a) M(a) CON STG 21-22')
cumm_df = create_one_year_or_cumm_data('RW-45-I-E(b) M(b ) CUMU TLL 22')

def get_data_for_branch(df,branch_code,category):
    df=df[df['BR CODE']==branch_code]
    df=df[df['category_new']==category]
    df=df[['Construction completed/ Under Construction','SUBSIDY_count','SUBSIDY_sum']]
    df.rename(columns = {'Construction completed/ Under Construction':'Particulars', 'SUBSIDY_count':'No. of Accounts','SUBSIDY_sum':'Total subsidy received from SBI(CNA)'}, inplace = True)
    df.index=df.Particulars
    return df
```

```python
from docx import Document
from docx.enum.text import WD_ALIGN_PARAGRAPH

document = Document()

def write_on_doc(branch_code,category):
    BRANCH_CODE=branch_code
    CATEGORY=category
    DETAIL="RW-45-I-M" if category=='MIG' else 'RW-45-I-E'
    document.add_heading("Branch Code: " + str(BRANCH_CODE), 0)

    p = document.add_paragraph('Annual End Use Certificate\n\n')

    document.add_paragraph(DETAIL)
    p2=document.add_paragraph(CATEGORY)
    p2.alignment = WD_ALIGN_PARAGRAPH.RIGHT

    # document.add_heading('Heading, level 1', level=1)
    document.add_paragraph("We certify that the construction of houses/flats for which the subsidy has been received and utilized is strictly \
    in terms of the guidelines issued by the Ministry of Housing & Urban Affairs, Government of India. ")

    document.add_paragraph('a. During the FY 2021-22\n')
    
    df_one_year=get_data_for_branch(one_year_df,BRANCH_CODE,CATEGORY)
    df1 = pd.DataFrame(columns=['Particulars', 'No. of Accounts', 'Total subsidy received from SBI(CNA)'], 
                       index=['Completed','Under Construction','Total'])
    df1['Particulars']=['Completed','Under Construction','Total']
    df1=df1.fillna(0)
    if len(list(df_one_year.index)) > 0:
        for key in df_one_year.index:
            # df1.loc[key,'Particulars']=df_one_year[key,'Particulars']
            # df1.loc[key,'No. of Accounts']=df_one_year[key,'No. of Accounts']
            # df1.loc[key,'Total subsidy received from SBI(CNA)']=df_one_year[key,'Total subsidy received from SBI(CNA)']
            df1.loc[key]=df_one_year.loc[key]
    # start table
    t1 = document.add_table(df1.shape[0]+1, df1.shape[1]) # One for header and 1 for total
    t1.style = 'Table Grid'

    # add the header rows.
    for j in range(df1.shape[-1]):
        t1.cell(0,j).text = df1.columns[j]

    # add the rest of the data frame
    for i in range(df1.shape[0]):
        for j in range(df1.shape[-1]):
            t1.cell(i+1,j).text = str(df1.values[i,j])
            
     #Total Row
    for j in range(1,df1.shape[-1]):
        t1.cell(df1.shape[0],j).text = str(sum(df1.values[:,j]))

    document.add_paragraph('\n\nb. Cumulative till 31-03-2022 \n')
    
    df_cumm=get_data_for_branch(cumm_df,BRANCH_CODE,CATEGORY)
    df2 = pd.DataFrame(columns=['Particulars', 'No. of Accounts', 'Total subsidy received from SBI(CNA)'], 
                       index=['Completed','Under Construction','Total'])
    df2['Particulars']=['Completed','Under Construction','Total']
    df2=df2.fillna(0)
    if len(list(df_cumm.index)) > 0:
        for key in df_cumm.index:
            # df2.loc[key,'Particulars']=df_cumm[key,'Particulars']
            # df2.loc[key,'No. of Accounts']=df_cumm[key,'No. of Accounts']
            # df2.loc[key,'Total subsidy received from SBI(CNA)']=df_cumm[key,'Total subsidy received from SBI(CNA)']
            df2.loc[key]=df_cumm.loc[key]
    # start table
    t2 = document.add_table(df2.shape[0]+1, df2.shape[1]) # One for Column and one for Total row
    t2.style = 'Table Grid'

    # add the header rows.
    for j in range(df2.shape[-1]):
        t2.cell(0,j).text = df2.columns[j]

    # add the rest of the data frame
    for i in range(df2.shape[0]):
        for j in range(df2.shape[-1]):
            t2.cell(i+1,j).text = str(df2.values[i,j])
    
    #Total Row
    for j in range(1,df2.shape[-1]):
        t2.cell(df2.shape[0],j).text = str(sum(df2.values[:,j]))
    

    document.add_paragraph("\n\nAs on 31-03-2022, under construction category, there were ____beneficiary loan accounts for which \
    consolidated utilization certificate has not been submitted to SBI,CC (CNA) for a period exceeding 36 months \
    from the date of the disbursement of the 1st instalment/tranche of the loan amount as per para E.6 of MoU \
    executed by LHO, PLI with SBI,CC (CNA). The details of the accounts are annexed. ")

    document.add_page_break()
    
for branch_code in all_branch_codes:
    write_on_doc(branch_code,category='EWS')
    write_on_doc(branch_code,category='MIG')
    
# save document
document.save("C:/Users/mithu/Downloads/output_docs/annual_end_use/out.docx")
```

```python
import pandas as pd
    
def create_annual_refund_data(excel_sheet_name):
    annual_refund=pd.read_excel("C:/Users/mithu/Desktop/mona study/RW 45 -ANNUAL CLOSING DATA  21-22 BANGALORE final.xlsx",skiprows=2,sheet_name=excel_sheet_name)
    annual_refund.columns=['CIRCLE', 'ACCOUNT_NO', 'NW', 'AO', 'RBO', 'BR_CODE', 'SUBSIDY',
       'P_FEES', 'TOTAL_SUBSIDY',
       'Category_Caste', 'Category',
       'SUBSIDY_RETURNED', 'P_FEES_RETURNED', 'TOTAL_SUBSIDY_RETURNED',
       'DATE OF SUBSIDY RETURNED DURING FY 21-22']
    annual_refund['category_new']=annual_refund['Category'].apply(lambda x: 'EWS' if x in ['EWS NEW','EWS OLD'] else 'MIG')
    annual_refund_pivot=annual_refund.groupby(['BR_CODE','category_new','Category_Caste']).agg({'TOTAL_SUBSIDY':['sum','count'],'SUBSIDY':'sum'
                                                ,'P_FEES':'sum' ,'SUBSIDY_RETURNED':['count','sum'],'P_FEES_RETURNED':'sum'
                                                ,'TOTAL_SUBSIDY_RETURNED':'sum'}).reset_index()
    annual_refund_pivot.columns=[x[0]+'_'+x[1] for x in annual_refund_pivot.columns]
    annual_refund_pivot=annual_refund_pivot[['BR_CODE_','category_new_', 'Category_Caste_', 
       'TOTAL_SUBSIDY_count', 'SUBSIDY_sum', 'P_FEES_sum','TOTAL_SUBSIDY_sum',
       'SUBSIDY_RETURNED_count', 'SUBSIDY_RETURNED_sum', 'P_FEES_RETURNED_sum',
       'TOTAL_SUBSIDY_RETURNED_sum']]
    annual_refund_pivot['Category_Caste_']=annual_refund_pivot['Category_Caste_'].apply(lambda x:Category_Caste[x])
    return annual_refund_pivot

annual_refund_pivot=create_annual_refund_data('RE-45-II-E DISB FY 21-22 GRAN')

def create_branch_data(df,branch_code,category):
    df=df[df['BR_CODE_']==branch_code]
    df=df[df['category_new_']==category]
    df.index=df['Category_Caste_']
    df=df[['Category_Caste_','TOTAL_SUBSIDY_count', 'SUBSIDY_sum', 'P_FEES_sum','TOTAL_SUBSIDY_sum',
       'SUBSIDY_RETURNED_count', 'SUBSIDY_RETURNED_sum', 'P_FEES_RETURNED_sum',
       'TOTAL_SUBSIDY_RETURNED_sum']]
    df.columns=['Purpose', 'No. of Accounts', 'Intersest Subsidy','Processing Fees',
                               'Total Subsidy','No. of Accounts','Interest Subsidy','Processing Fees',
                               'Total Subsidy']
    return df
```

```python
from docx import Document
from docx.enum.text import WD_ALIGN_PARAGRAPH
from docx.shared import Pt

document = Document()

def write_on_doc(branch_code,category):
    BRANCH_CODE=branch_code
    CATEGORY=category
    DETAIL="RW-45-II-M" if category=='MIG' else 'RW-45-II-E'
    document.add_heading("Branch Code: " + str(BRANCH_CODE), 0)
    
    p2=document.add_paragraph(DETAIL)
    p2.alignment = WD_ALIGN_PARAGRAPH.RIGHT
    
    p = document.add_paragraph('Annual Utilization & Refund Certificate\n\n')
    p.alignment = WD_ALIGN_PARAGRAPH.CENTER
    p.bold = True
    p.add_run(CATEGORY).bold=True

    # document.add_heading('Heading, level 1', level=1)
    p3=document.add_paragraph('''Implementation of Pradhan Mantri Awas Yojana (PMAY) - Credit Linked Subsidy Scheme (CLSS)
        for MIG by State Bank of India, CC 
        (Central Nodal Agency) \n''')
    p3.alignment = WD_ALIGN_PARAGRAPH.CENTER

    p4=document.add_paragraph('Statement on Interest Subsidy Claimed and Refund Made during FY 2021-22 by')
    p4.alignment = WD_ALIGN_PARAGRAPH.CENTER
    
    df_utilization=create_branch_data(annual_refund_pivot,BRANCH_CODE,CATEGORY)
    df1 = pd.DataFrame(columns=['Purpose', 'No. of Accounts', 'Intersest Subsidy','Processing Fees',
                               'Total Subsidy','No. of Accounts','Interest Subsidy','Processing Fees',
                               'Total Subsidy'], 
                       index=['General','SC','ST','OBC','Total'])
    df1['Purpose']=['General','SC','ST','OBC','Total']
    df1=df1.fillna(0)
    if len(list(df_utilization.index)) > 0:
        for key in df_utilization.index:
            df1.loc[key]=df_utilization.loc[key]
    # start table
    t1 = document.add_table(df1.shape[0]+1, df1.shape[1]) # One for header and 1 for total
    t1.style = 'Table Grid'
    # t1.autofit = True

    # add the header rows.
    for j in range(df1.shape[-1]):
        t1.cell(0,j).text = df1.columns[j]

    # add the rest of the data frame
    for i in range(df1.shape[0]):
        for j in range(df1.shape[-1]):
            t1.cell(i+1,j).text = str(df1.values[i,j])
            
     #Total Row
    for j in range(1,df1.shape[-1]):
        t1.cell(df1.shape[0],j).text = str(sum(df1.values[:,j]))
        
    for row in t1.rows:
        for cell in row.cells:
            paragraphs = cell.paragraphs
            paragraph = paragraphs[0]
            run_obj = paragraph.runs
            run = run_obj[0]
            font = run.font
            font.size = Pt(9)
    
    document.add_paragraph("\n\nWe certify that the amounts stated above have been received and utilized strictly in terms of the guidelines issued by the Ministry of Housing & Urban Affairs, Government of India.")
    document.add_paragraph("\nWe also certify that the above claims made as true and correct and declare that the same is as per the records of the books of the Bank.")
    document.add_page_break()
    
for branch_code in all_branch_codes:
    write_on_doc(256,category='EWS')
    write_on_doc(256,category='MIG')
    
# save document
document.save("C:/Users/mithu/Downloads/output_docs/annual_utilization and refund/out.docx")
```
